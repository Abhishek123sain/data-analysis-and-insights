```python
# Loading the required libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Loading IPL matches Dataset
ipl = pd.read_csv(r"C:\Users\DELL\Desktop\EDA on IPL\matches (1).csv")
```

```python
#First five rows of Dataset
ipl.head()
```

```python
#Shape of Dataset
ipl.shape
```

```python
# Most man of the match
ipl['player_of_match'].value_counts()
```

```python
# Top 10 most man of the match
ipl['player_of_match'].value_counts()[0:10]
```

```python
# Top 5 most man of the match
ipl['player_of_match'].value_counts()[0:5]
```

```python
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
# plot a bar-plot for the top 5 players with most man of the match]
plt.figure(figsize = (9, 5))
plt.bar(ipl['player_of_match'].value_counts()[0:10].keys(),ipl['player_of_match'].value_counts()[0:10],color = 'r')
plt.show()
```

```python
# frequency of result columns
ipl['result'].value_counts()
```

```python
# The number of toss win
ipl['toss_winner'].value_counts()
```

```python
# The recordd where a team batting first
batting_first = ipl[ipl['win_by_runs'] != 0]
```

```python
# first five rows of batting first tean
batting_first.head()
```

```python
# MAking histogram of batting first team
plt.figure(figsize = (5, 3))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.show()
```

```python
 # Winner after batting first
batting_first['winner'].value_counts()
```

```python
# Bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize = (5,3))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()), list(batting_first['winner'].value_counts()[0:3]), color = ['blue','yellow','red'])
plt.show()
```

```python
# Making a Pie Chart
plt.figure(figsize = (5,5))
plt.pie(list(batting_first['winner'].value_counts()), labels = list(batting_first['winner'].value_counts().keys()), autopct = '%0.1f%%')
plt.show()
```

```python
# team won after batting second
batting_second = ipl[ipl['win_by_wickets'] != 0]
```

```python
# Head of batting second 
batting_second.head()
```

```python
# Histogram for batting second team
plt.figure(figsize = (5, 3))
plt.hist(batting_second['win_by_wickets'], bins = 30)
plt.show()
```

```python
# Bar-{Plot for top 3 team win after batting second
plt.figure(figsize = (5,3))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()), list(batting_second['winner'].value_counts()[0:3]),color = ['purple','blue','orange'])
plt.show()
```

```python
# Making pie chart
plt.figure(figsize = (5,5))
plt.pie(list(batting_second['winner'].value_counts()), labels = list(batting_second['winner'].value_counts().keys()) , autopct = '%0.1f%%')
plt.show()
```
